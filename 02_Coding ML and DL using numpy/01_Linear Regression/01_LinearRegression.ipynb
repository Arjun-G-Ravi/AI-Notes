Dataset

```python
from sklearn import datasets
ds = datasets.load_diabetes()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ds['data'], ds['target'])
```

Data Visualisation

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(5, 2,figsize=(10,15))
ct = 0
for i in range(5):
    for j in range(2):
        axs[i,j].scatter(X_train[:,ct], y_train)
        axs[i,j].set_title('Subplot 1')
        ct += 1
plt.tight_layout()
```

```python
plt.hist(y_train,bins=50)
plt.show()
```

# Linear Regression

Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
print(f'Weight:{model.coef_}\nBias:{model.intercept_}')
```

Evaluation

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

Underfitting :(

# Ridge - to overcome overfitting

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=.7).fit(X_train, y_train)  # alpha is the regularising term to prevent 
```

```python
ridge.score(X_train,y_train)
```

```python
ridge.score(X_test,y_test)
```

# Lasso - another regularisation

reduces the feature usage

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=.001, max_iter=100000).fit(X_train, y_train)
```

```python
lasso.score(X_train,y_train)
```

```python
lasso.score(X_test,y_test)
```

```python
import numpy as np
print("Number of features used: {}".format(np.sum(lasso.coef_ != 0)))
```

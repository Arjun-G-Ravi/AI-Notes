Dataset

```python
from sklearn import datasets
ds = datasets.load_diabetes()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(ds['data'], ds['target'])
```

Data Visualisation

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(5, 2,figsize=(10,15))
ct = 0
for i in range(5):
    for j in range(2):
        axs[i,j].scatter(X_train[:,ct], y_train)
        axs[i,j].set_title('Subplot 1')
        ct += 1
plt.tight_layout()
```

```python
plt.hist(y_train,bins=50)
plt.show()
```

Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
print(f'Weight:{model.coef_}\nBias:{model.intercept_}')
```

Inference

```python
model.predict([[ 0.03807591,0.05068012,-0.02991782,-0.04009893,-0.03321588,-0.02417372, -0.01026611,-0.00259226,-0.01290868,0.00306441]])
```

Evaluation

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
